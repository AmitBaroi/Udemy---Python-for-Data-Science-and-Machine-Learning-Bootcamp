# Setup

```python
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

___Initializing notebook mode___: Allows interactive plots to be displayed in Jupyter Notebooks.

```python
init_notebook_mode(connected=True)
```

# Data

```python
data = dict(type = 'choropleth',
           locations = ['LA', 'CA', 'NY', 'TX'],
           locationmode = 'USA-states',
           colorscale = 'Blackbody',
           text = ['text 1', 'text 2', 'text 3', 'text 4'],
           z = [1.0, 2.0, 3.0, 2.5],
           colorbar = {'title':'Colorbar title!'})
```

# Layout

```python
layout = dict(geo={'scope':'usa'})
```

# Plotly graph object

```python
choromap = go.Figure(data=[data], layout=layout)
```

# Plotting

```python
iplot(choromap)
```

# Real world data

```python
import pandas as pd
df = pd.read_csv("2011_US_AGRI_Exports")
df.head()
```

## Preparing data
Available `colorscale`s:

'Greys', 'YlGnBu', 'Greens', 'YlOrRd', 'Bluered', 'RdBu', 'Reds', 'Blues', 'Picnic', 'Rainbow', 'Portland', 'Jet', 'Hot', 'Blackbody', 'Earth', 'Electric', 'Viridis','Cividis'

```python
data = dict(type='choropleth',
            colorscale = 'Reds',
            locations = df['code'],
            z = df['total exports'],
            locationmode = 'USA-states',
            text = df['text'],
            marker = dict(line = dict(color = 'rgb(255,255,255)',width = 2)),
            colorbar = {'title':"Millions USD"}
            ) 
```

## Preparing layout

```python
layout = dict(title = '2011 US Agriculture Exports by State',
              geo = dict(scope='usa',
                         showlakes = True,
                         lakecolor = 'rgb(85,173,240)')
             )
```

## Preparing figure object

```python
choromap = go.Figure(data=[data], layout=layout)
```

```python
iplot(choromap)
```

# Geographical Plotting on Global Scale
World GDP dataset 2014

```python
df = pd.read_csv('2014_World_GDP')
df.head()
```

```python
data = dict(type='choropleth',
           locations=df['CODE'],
           z=df['GDP (BILLIONS)'],
           text=df['COUNTRY'],
           colorbar={'title':'GDP in Billions USD'})
```

```python
layout = dict(title='World GDP by Country - 2014',
             geo=dict(showframe=False,
                     projection={'type':'mercator'}))
```

```python
choromap = go.Figure(data=[data], layout=layout)
iplot(choromap)
```
